## 선택 정렬
- 가장 간단한 방법
- 가장 작은 숫자를 찾아서 맨 앞으로 보낸다.
- 그 다음 작은 숫자를 찾아서 두번째로 보낸다.
- 반복한다

```python
# 파이썬 기본함수를 이용한 구현
%%time
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

sort_array = []
while array:
#     print(i)
    sort_array.append(min(array))
    array.remove(min(array))
    
sort_array
```

```python
%%time
# 책에서의 선택정렬 구현

array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

for i in range(len(array)):
    min_index = i
    for j in range(i+1, len(array)):
        if array[j] < array[min_index]:
            min_index = j     # 이렇게 for을 i+1부터 끝까지 반복하면 min_index는 결국 제일 작은 값의 원래 index로 바뀐다
            
    array[i], array[min_index] = array[min_index], array[i]  # 찾은 최소값을 처음값과 swap한다.
    
print(array)
        
```

이 경우 시간 복잡도는 $O(N^2)$ 이다.

## 삽입 정렬
- 필요할 때만 위치를 바꾸므로 데이터가 거의 정렬되어 있을때 훨씬 효율적
- 선택 정렬보다 구현 난이도가 높지만 더 효율적
- 두번째 데이터부터 시작하여 왼쪽 데이터들과 비교하여 알맞은 위치에 '삽입'함

```python
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

array.insert(1,3) # 위치, 숫자
array
```

```python
array.remove(8)
array
```

```python
%%time
# 삽입정렬 스스로 구현
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

for i in range(1,len(array)):
    value = array[i] # 선택한 값
    print(i, value)
    for j in range(i):  # jj i 
        if array[j] >= value: # 좌측부터 훑으면서 큰값인곳을 찾아
            array.remove(value)  # 원래 위치에서는 제거
            array.insert(j,value)  # 그 큰값 왼쪽에 삽입한다.
            break  # 더 정렬할 필요 없으므로 break
            

array
```

```python
%%time
# 삽입정렬 책대로 구현
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]

for i in range(1, len(array)):
    
    for j in range(i-1, -1, -1): # 왼쪽값들중 오른쪽부터 끝까지
        if array[j] > array[j+1]:
            array[j+1], array[j] = array[j], array[j+1]
            
            
array
```

## 퀵 정렬
- **가장 많이 사용되는 알고리즘**
- 대부분의 언어에서 정렬 라이브러리의 근간
- '기준(피벗) 데이터를 설정하고 그 기준보다 큰 데이터와 작은 데이터의 위치를 바꾼다'
- **삽입 정렬과 반대로, 정렬이 거의 다 되어 있을 경우에 느려진다**

```python
# 혼자 먼저 구현
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]
print(array)
def quick(arr):
    pivot = arr[0]
    
    if len(arr) == 1:
        return arr
    
    while True:  
        left, right = None, None
        for i in range(1,len(arr)):  
            if arr[i] > pivot: # pivot보다 큰 왼쪽 데이터
                left = arr[i]
                break
            
        for j in range(1, len(arr)):
            if arr[-j] < pivot: # pivot 보다 작은 오른쪽 데이터
                right = arr[-j]
                break
                
#         pdb.set_trace()
        
        
        
        
        print(left, right)
        # 두 데이터를 찾으면 서로 위치를 바꾼다
        
        if right is None: #오른쪽에 큰값만 존재
            print('오른쪽큰값만존재,',arr)
            return [pivot] + quick(arr[-j+1:])
        
        
        # 순서 섞이면 - pivot과 오른쪽 작은값을 서로 바꾸고 재귀 형태로 리턴
        elif i+j > len(arr):  
            arr[0], arr[-j] = arr[-j], arr[0]
            print('순서섞임', arr)
            return quick(arr[0:-j]) + [pivot] + quick(arr[-j+1:])
        
        elif left is not None:  
            arr[i], arr[-j] = right, left
            print('반복',arr)
            
        
    

quick(array)
```

포기... 책에 나온 소스코드로 학습하기

```python
# 퀵 정렬 소스코드 1
array = [5, 7, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort(array, start, end):
    if start >= end: # 원소가 1개일 경우 종료
        return
    pivot = start # 피벗은 첫번째 원소
    
    # 탐색을 처음에 정하여 하나씩 전진시킴
    left = start + 1  # 두번째부터 시작
    right = end
    
    while left <= right:
        # 피벗보다 큰 데이터를 찾을때까지 반복
        while left <= end and array[left] <= array[pivot]:
            left += 1
        
        # 피벗보다 작은 데이터를 찾을 때까지 반복
        while right > start and array[right] >= array[pivot]:
            right -= 1
        
        '''while 끝나면 찾았다는 뜻.
        처음에 시작,도착을 정해놨기 때문에 
        리스트가 2개여도 
        여전히 작동한다.'''
        
        if left > right: # 엇갈렸다면 작은 데이터와 피벗을 교체 - while 종료됨  
            array[right], array[pivot] = array[pivot], array[right]
        
        else:  # 안 엇갈렸으므로 다시 while 문으로 감
            array[left], array[right] = array[right], array[left]
    
    # 분할 이후 왼쪽 부분과 오른쪽 부분에서 각각 정렬 수행
    quick_sort(array, start, right-1)
    quick_sort(array, right+1, end)
    
quick_sort(array, 0, len(array) - 1)
array
```

```python
%%time
# 안보고 구현

array = [5, 7, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort(array, start, end):
    
    pivot = start
    left = start+1
    right = end
    
    # 길이 1인 경우는 종료
    if start >= end:
        return
    
    while left <= right:
        
        # 값 찾을때까지 반복하여 left, right를 이동
        while left <= end and array[pivot] > array[left]:
            left += 1
            
        while right > start and array[pivot] < array[right]:
            right -= 1
            
        
        if left > right: # 순서 바뀌는 경우 - pivot 과 right 를 교체
            array[pivot], array[right] = array[right], array[pivot]
            
        else: # 안바뀐 경우 - left 와 right 을 교체하고 다시 while문 
            array[left], array[right] = array[right], array[left]
            
    # 재귀 활용 - pivot 좌우 파트에서 다시 실행
    quick_sort(array, start, right-1)
    quick_sort(array, right+1, end)
    
quick_sort(array, 0, len(array)-1)
array
 
```

```python
# 퀵 정렬 소스코드 2
array = [5, 7, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort(array):
    # 리스트가 하나 이하의 원소를 담고 있다면 종료
    if len(array) <= 1:
        return array
    
    pivot = array[0] 
    tail = array[1:]
    
    left_side = [x for x in tail if x <= pivot]
    right_side = [x for x in tail if x > pivot]
    
    # 분할 이후 왼쪽 / 오른쪽 부분에서 각각 정렬을 수행하고, 전체 리스트를 반환
    return quick_sort(left_side) + [pivot] + quick_sort(right_side)
```

```python
%%time
# 퀵 정렬 소스코드 2 - 직접구현
array = [5, 7, 9, 0, 3, 1, 6, 2, 4, 8]

def quick_sort(array):
    
    
    if len(array) <= 1:
        return array
    
    pivot = array[0]
    tail = array[1:]
    
    pivot_left = [x for x in tail if x <= pivot]
    pivot_right = [x for x in tail if x > pivot]
   
    # 재귀 이용 - 양쪽 영역에서 다시 정렬 수행
    return quick_sort(pivot_left) + [pivot] + quick_sort(pivot_right)

quick_sort(array)
```

## 계수 정렬
- 앞의 정렬 방식들은 모두 비교하며 정렬하는 방식이지만 계수 정렬은 이와 다르다
- 정수형 등 discrete 한 데이터에 한해 적용 가능하고, 속도가 퀵 정렬과 더불어 가장 빠른 방법이다.
- 길이만큼의 정렬된 배열을 정의하고 인덱싱해서 count 한다. 그다음 처음부터 순서대로 읽는다.
- 시간 복잡도는 $O(N+K)$ 이다.
- 비효율성 초래 가능성 : 그리드는 커야 하는데 실질적인 데이터는 적은 경우 (0, 99999만 존재하는 배열을 예로 들 수 있음.)

```python
%%time
array = [7,5,9,0,3,1,6,2,9,1,4,8,0,5,2]

def num_sort(array):
    sort_arr = [0]*(max(array) - min(array) + 2)
    for num in array:
        sort_arr[num] += 1
        
    sort_ans = []
    for i in range(1,len(sort_arr)):
        for _ in range(sort_arr[i]):
            sort_ans.append(i)
    return sort_ans

num_sort(array)
        
    
```

## 코팅 테스트에서 정렬 알고리즘이 사용되는 경우
1. 정렬 라이브러리로 풀 수 있는 문제  
2. 정렬 알고리즘 원리에 대해서 물어보는 문제  
: 선택 정렬, 삽입 정렬, 퀵 정렬 등의 원리를 알고 있어야 적용 가능하다
3. 더 빠른 정렬이 필요한 문제  
: 계수 정렬 등의 다른 정렬 알고리즘을 사용하거나 기존 알고리즘의 구조적 개선이 필요하다.

## 실전 문제 - 두 배열의 원소 교체

```python
N, K = map(int, input().split())
A = [int(a) for a in input().split()]
B = [int(b) for b in input().split()]

A = sorted(A)
B = sorted(B, reverse=True)
for i in range(N):
    print(i)
    if A[i] < B[i]:
        A[i], B[i] = B[i], A[i]
    K -= 1
    if K == 0:
        break
        
print(A)
print(sum(A))
```
